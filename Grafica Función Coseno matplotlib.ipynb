Grafica de Serie de Taylor


```python
import numpy as np
import matplotlib.pyplot as plt
x= np.linspace(0,2*np.pi,100)
y=np.cos(x)
plt.xlabel("x")
plt.ylabel("cos(x)")
plt.title("Gráfica: Función Coseno")
plt.plot(x,y)
plt.show()
```

## Serie de McLaurin de cos(x) de orden 6

La serie de *McLaurin* de f(x) = \cos(x) se expresa como:

$$
\cos(x) = \sum_{n=0}^{\infty} \frac{(-1)^n x^{2n}}{(2n)!}
$$

Aproximando hasta el *término de orden 6*:

$$
\cos(x) \approx 1 - \frac{x^2}{2!} + \frac{x^4}{4!} - \frac{x^6}{6!}
$$

Sustituyendo los valores de los factoriales:

$$
\cos(x) \approx 1 - \frac{x^2}{2} + \frac{x^4}{24} - \frac{x^6}{720}
$$
