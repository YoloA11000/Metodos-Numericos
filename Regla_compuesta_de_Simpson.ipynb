```python
import numpy as np
import matplotlib.pyplot as plt

# ---------------------------------------------------
# Función a integrar
# f(x) = x^4 - 2x + 1
# ---------------------------------------------------
def f(x):
    return x**4 - 2*x + 1

# ---------------------------------------------------
# Parámetros:
#   a: límite inferior de integración
#   b: límite superior de integración
#   n: número de subintervalos (debe ser PAR)
# Devuelve:
#   Aproximación de la integral definida de f(x)
# ---------------------------------------------------
def regla_simpson_compuesta(a, b, n):
    if n % 2 != 0:
        raise ValueError("El número de subintervalos 'n' debe ser par.")

    h = (b - a) / n  # Tamaño de subintervalo
    suma = f(a) + f(b)  # Suma de los extremos

    # Sumar los valores intermedios con los coeficientes 4 y 2
    for i in range(1, n):
        x = a + i * h
        if i % 2 == 0:
            suma += 2 * f(x)
        else:
            suma += 4 * f(x)

    return (h / 3) * suma

# -------------------------------
# Parámetros del Ejemplo 1
# Intervalo [a, b] = [0, 2]
# Subintervalos: n = 10
# -------------------------------
a = 0
b = 2
n = 10

# Cálculo de la integral usando la regla compuesta de Simpson
resultado = regla_simpson_compuesta(a, b, n)

# Mostrar el resultado con 6 cifras decimales
print(f"Resultado aproximado con la regla compuesta de Simpson: {resultado:.6f}")

# -------------------------------
# Gráfica de la función y sombreado
# -------------------------------
x_vals = np.linspace(a, b, 100)  # Valores de x para graficar
y_vals = f(x_vals)               # Evaluar f(x) para cada x

plt.plot(x_vals, y_vals, label='f(x) = x⁴ - 2x + 1', color='blue')
plt.title('Regla Compuesta de Simpson — Ejemplo 1 (Burden)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()

# Sombrear el área bajo la curva (integral aproximada)
plt.fill_between(x_vals, y_vals, color='orange', alpha=0.3)
plt.show()
```
